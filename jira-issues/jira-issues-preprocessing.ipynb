```python
import pandas as pd
import numpy as np
import re
```

```python
utterances_df = pd.read_csv("data/Etisalat Jira 2019-05-23T11_54_29+0400.csv", encoding = 'utf-8')
```

```python
utterances_df.shape
```

```python
utterances_df.head(10)
```

```python
utterances_ar_df = utterances_df[["Issue key", "Issue id", "Summary", "Description"]]
```

```python
utterances_ar_df["Summary_ar"] = utterances_ar_df["Summary"]
utterances_ar_df["Summary_ar"] = utterances_ar_df["Summary_ar"].map(lambda line: re.sub("[^\u0621-\u06FF0-9/]+", ' ', str(line)))
```

```python
utterances_ar_df.head()
```

```python
utterances_ar_df.shape
```

```python
# Mark all empty arabic summary as nan
utterances_ar_df = utterances_ar_df.replace(r'^\s*$', np.nan, regex=True)
```

```python
# Create new df for English only summary
utterances_en_df = utterances_ar_df[utterances_ar_df.Summary_ar.isnull()]
```

```python
utterances_en_df.shape
```

```python
utterances_en_df.head()
```

```python
#Remove empty arabic summary values
utterances_ar_df = utterances_ar_df.dropna(subset=["Summary_ar"])
```

```python
utterances_ar_df.shape
```

```python
#Prepare the english only summary file to include arabic text from Description
```

```python
utterances_en_df["Description_ar"] = utterances_en_df["Description"]

#utterances_en_df["Description_ar"] = utterances_en_df["Description_ar"].map(lambda line: re.sub("[^\u0621-\u06FF0-9/]+", ' ', str(line)))

#utterances_en_df["Description_ar"] = utterances_en_df["Description_ar"].str.replace('\d+', '')

utterances_en_df.Description_ar = utterances_en_df.Description_ar.str.replace('[^\u0621-\u06FF0-9/]+', ' ').str.replace('\d+', ' ').str.replace('[^\w\s]',' ').str.strip()
```
